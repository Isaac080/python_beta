```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
%pylab inline
%matplotlib inline
```

What kind of curve in the $xy$-plane is parametrized by:  
$$x=3sin2t, y=cos2t, 0\leq t\leq \pi$$

```python
t=np.linspace(0,np.pi,100)
x=3*sp.sin(2*t)
y=sp.cos(2*t)
plt.plot(x,y)
```

```python
t=np.linspace(0,1,100)
```

```python
y=1.0-t**2
plt.plot(t,y)
```

```python
y=np.e**-t
plt.plot(t,y)
```

```python
y=2*np.e**(2*t)-np.e**t
plt.plot(t,y)
```

```python
y=4+sp.cos(np.pi*t)
plt.plot(t,y)
```

```python
y=t**2-1
plt.plot(t,y)
```
