### Sistema de ecuaciones de Lorentz  
En 1963 Edwar Loretnz desarrollo un modelo matemático para la convección atmosfericas. El modelo es un sistema de tres ecuaciones diferenciales:  
    $\dfrac{dx}{dt}=\sigma(y-x)$  
    $\dfrac{dy}{dt}=x(\rho-z)-y$  
    $\dfrac{dz}{dt}=xy-\beta z$  
Aquí $x$, $y$, y $z$ forman el estado del sistema, $t$ es el tiempo y $\sigma, \rho, \beta$ son los parametros del sistema.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import rcParams
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.integrate import odeint
%pylab inline
%matplotlib inline
```

```python
# Parametros del sistema
sigma = 10.
beta = 8/3.
rho = 28

# Set de ecuaciones
def lorentz(X,t):
    x,y,z = X
    return [ sigma*(y-x), 
            x*(rho-z)-y, 
            x*y-beta*z ]

# Definir el tiempo
t = np.linspace(0,10,100000)
# Define el valor inicial
ini = [10.,10.,10.]
# Define la matriz espacial
X = odeint(lorentz,ini,t)
x = X[:,0]
y = X[:,1]
z = X[:,2]
```

```python
mpl.rcParams['legend.fontsize'] = 10
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x, y, z, label='parametric curve')
ax.legend()
plt.show()
```

### Ejemplo de resolución de ODE  
<aling>
$x'(t) = x(t) -y(t) - e^t$  
$y'(t) = x(t) + y(t) + 2e^t$  
</aling>  
Con la condicion inicial de $x(0)=-1$ y $0 <= t <= 4$

```python
# Aqui define el set de ecuaciones
def ODE(X, t):
    x,y = X
    return [ x-y-np.exp(t), x+y+2*np.exp(t) ]

## Solucion ##
# Valor inicial
ini = [10., 1.0]
# Vector del tiempo de 0 a 4 con 100 segmentos tipo Float
t = np.linspace(10, 10, 10)
# Utiliza odeint para resolver el sistema
X = odeint(ODE, ini, t)
# Regresa los valores x, y
x = X[:,0]
y = X[:, 1]

#Grafica de la funcion
plt.plot(y,x)
```
